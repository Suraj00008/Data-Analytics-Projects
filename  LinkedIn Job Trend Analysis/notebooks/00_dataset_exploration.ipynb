```python
keywords = [
    "data analyst",
    "data scientist",
    "data engineer",
    "machine learning engineer",
    "business analyst",
    "analytics engineer"
]

pattern = "|".join(keywords)

data_jobs = postings[
    postings["title"].str.lower().str.contains(
        pattern,
        na=False
    )
].copy()

print(data_jobs.shape)
```

```python
skills_list = [
    "python",
    "sql",
    "excel",
    "power bi",
    "tableau",
    "aws",
    "azure",
    "gcp",
    "snowflake",
    "kafka",
    "spark",
    "hadoop",
    "etl",
    "machine learning",
    "deep learning",
    "statistics",
    "r",
    "pandas",
    "numpy"
]
```

```python
from collections import Counter

skill_counter = Counter()

for description in data_jobs["description"].dropna():
    description = description.lower()

    for skill in skills_list:
        if skill in description:
            skill_counter[skill] += 1
```

```python
top_skills = (
    pd.DataFrame(
        skill_counter.items(),
        columns=["skill", "count"]
    )
    .sort_values(by="count", ascending=False)
)

top_skills.head(15)
```

```python
import matplotlib.pyplot as plt

top_10 = top_skills.head(10)

plt.figure(figsize=(10, 6))

plt.bar(top_10["skill"], top_10["count"])

plt.xticks(rotation=45)
plt.title("Top Technical Skills in Data Jobs")
plt.xlabel("Skill")
plt.ylabel("Job Count")

plt.tight_layout()
plt.show()
```

```python
from collections import Counter

def get_top_skills(job_keyword, top_n=10):

    role_data = data_jobs[
        data_jobs["title"].str.lower().str.contains(
            job_keyword,
            na=False
        )
    ]

    counter = Counter()

    for description in role_data["description"].dropna():
        description = description.lower()

        for skill in skills_list:
            if skill in description:
                counter[skill] += 1

    return pd.DataFrame(
        counter.items(),
        columns=["skill", "count"]
    ).sort_values(
        by="count",
        ascending=False
    ).head(top_n)
```

```python
analyst_skills = get_top_skills("data analyst")
scientist_skills = get_top_skills("data scientist")
engineer_skills = get_top_skills("data engineer")

print("Data Analyst")
display(analyst_skills)

print("Data Scientist")
display(scientist_skills)

print("Data Engineer")
display(engineer_skills)
```

```python
analyst_skills.to_csv("../data/analyst_skills.csv", index=False)
scientist_skills.to_csv("../data/scientist_skills.csv", index=False)
engineer_skills.to_csv("../data/engineer_skills.csv", index=False)
```

```python
salary_analysis = (
    data_jobs[
        ["title", "max_salary", "min_salary"]
    ]
    .dropna(subset=["max_salary", "min_salary"])
)

salary_analysis["avg_salary"] = (
    salary_analysis["max_salary"] +
    salary_analysis["min_salary"]
) / 2

salary_analysis.head()
```

```python
top_salary_roles = (
    salary_analysis.groupby("title")["avg_salary"]
    .mean()
    .sort_values(ascending=False)
    .head(15)
    .reset_index()
)

top_salary_roles
```

```python
top_salary_roles = top_salary_roles[
    top_salary_roles["avg_salary"] < 500000
]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.barh(
    top_salary_roles["title"],
    top_salary_roles["avg_salary"]
)

plt.title("Highest Paying Data Roles")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```
